```python
# 1. Traffic Sources Pie Chart
traffic_counts = data["Traffic Source"].value_counts()
traffic_labels = traffic_counts.index

plt.figure(figsize=(8, 6))
plt.pie(traffic_counts, labels=traffic_labels, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Traffic Sources Distribution")
#plt.show()
plt.savefig("Traffic Sources Distribution.png")
```

```python
# 2. Conversion Rates by Traffic Source (Bar Chart)
conversion_rates = data.groupby("Traffic Source")["Conversion Rate"].mean().sort_values()

plt.figure(figsize=(8, 5))
sns.barplot(x=conversion_rates.values * 100, y=conversion_rates.index, palette="viridis")
plt.xlabel("Conversion Rate (%)")
plt.ylabel("Traffic Source")
plt.title("Conversion Rate by Traffic Source")
#plt.show()
plt.savefig("Conversion Rate by Traffic Source.png")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Conversion Rates by Traffic Source (Bar Chart)
conversion_rates = data.groupby("Traffic Source")["Conversion Rate"].mean().sort_values()

plt.figure(figsize=(10, 6))  # Increase figure size
ax = sns.barplot(x=conversion_rates.values * 100, y=conversion_rates.index, palette="viridis")

# Add value labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f%%", padding=5)  

plt.xlabel("Conversion Rate (%)")
plt.ylabel("Traffic Source")
plt.title("Conversion Rate by Traffic Source")

plt.savefig("Conversion Rate by Traffic Source 2.png")
# plt.show()
```
