```python
# create short titles
def create_short_title(title):
    words_to_remove = ["includes", "set of", "features", "for", "with", "|", ","]
    short_title = title
    for word in words_to_remove:
        short_title = short_title.replace(word, "")
    
    short_title = " ".join(short_title.split())  # Remove extra spaces
    return short_title[:50]  # Limit to 50 characters


## Apply the function to the TITLE column

#dropped['short_title'] = dropped['title'].apply(create_short_title)
```

```python
import re
```

```python
# function to create a short_title
def generate_short_title(title):
    # Remove unnecessary words and special characters
    title = re.sub(r'\b(set of|includes|features|for|with|&|and|,)\b', '', title, flags=re.IGNORECASE)
    # Keep only useful words. set Limit to only 60 characters
    short_title = " ".join(title.split()[:6]) # Keep only the first six words
    return short_title.strip()



## Apply the function to the TITLE column
dropped['short_title'] = dropped['title'].apply(generate_short_title)
```

```python
# Function to remove anomalies based on IQR
def remove_anomalies(df, column):
    Q1 = dropped[column].quantile(0.25)
    Q3 = dropped[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return dropped[(dropped[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply anomaly removal to a numerical column (replace 'price' with your column)
dropped = remove_anomalies(dropped, 'productid')
```
